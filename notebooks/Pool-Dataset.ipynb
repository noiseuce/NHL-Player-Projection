## Join by Name

```python
nhl13.shape

nhl14.shape

nhl13_14 = nhl14.join(nhl13)
nhl13_14 = pd.concat([nhl13_14, nhl13], sort=False)
nhl13_14 = nhl13_14[~nhl13_14.index.duplicated(keep='first')]

nhl13_14.shape

nhl14_15 = nhl15.join(nhl13_14)
nhl14_15 = pd.concat([nhl14_15, nhl13_14], sort=False)
nhl14_15 = nhl14_15[~nhl14_15.index.duplicated(keep='first')]

nhl14_15.shape

nhl15_16 = nhl16.join(nhl14_15)
nhl15_16 = pd.concat([nhl15_16, nhl14_15], sort=False)
nhl15_16 = nhl15_16[~nhl15_16.index.duplicated(keep='first')]

nhl15_16.shape

nhl16_17 = nhl17.join(nhl15_16)
nhl16_17 = pd.concat([nhl16_17, nhl15_16], sort=False)
nhl16_17 = nhl16_17[~nhl16_17.index.duplicated(keep='first')]

nhl16_17.shape

nhl16_17pp = nhl17pp.join(nhl16_17)
nhl16_17pp = pd.concat([nhl16_17pp, nhl16_17], sort=False)
nhl16_17pp = nhl16_17pp[~nhl16_17pp.index.duplicated(keep='first')]

nhl16_17pp.shape

nhl17_18pp = nhl18pp.join(nhl16_17pp)
nhl17_18pp = pd.concat([nhl17_18pp, nhl16_17pp], sort=False)
nhl17_18pp = nhl17_18pp[~nhl17_18pp.index.duplicated(keep='first')]

nhl17_18pp.shape

nhl17_18 = nhl18.join(nhl17_18pp)
nhl17_18 = pd.concat([nhl17_18, nhl17_18pp], sort=False)
nhl17_18 = nhl17_18[~nhl17_18.index.duplicated(keep='first')]

nhl17_18.shape

nhl18_19pp = nhl19pp.join(nhl17_18)
nhl18_19pp = pd.concat([nhl18_19pp, nhl17_18], sort=False)
nhl18_19pp = nhl18_19pp[~nhl18_19pp.index.duplicated(keep='first')]

nhl18_19pp.shape

nhl = nhl19.join(nhl18_19pp)
nhl = pd.concat([nhl, nhl18_19pp], sort=False)
nhl = nhl[~nhl.index.duplicated(keep='first')]
nhl.index.name = 'Name'

nhl.shape
```

## Filter columns

```python
nhl['Age_19'] = nhl.apply(lambda x: x.Age_18+1, axis=1)
nhl['TOI/GP_19'] = nhl.apply(lambda x: x.TOI_19/x.GP_19, axis=1)
nhl['TOI/GP_19pp'] = nhl.apply(lambda x: x.TOI_19pp/x.GP_19pp, axis=1)
nhl = nhl.rename(index=str, columns={"P_19": "PTS_19", "iSh%_19": "SH%_19", "ZSR_19": "ZS%_19", "GS/60_19": "GS/G_19"})

# add 17 et 18 PPTOI des xls

# PAX: Points above expectations, based on previous average career points per game
# GC: Goals Created, which is half-point for a goal, and a variable amount depending on the team's assists per goal
# SCF: Team scoring chances while this player was on the ice
# WAR: Wins above replacement, a catch-all stat that measures a player's offensive contributions in terms of wins
# ZS%: Zone start percentage
# GS: The player's combined game score, quality of games played, averaged per game

nhl = nhl[['Last Name_19', 'First Name_19', 'Team_19', 'Age_19', 'Position_19', \
           'GP_13', 'GP_14', 'GP_15', 'GP_16', 'GP_17', 'GP_18', 'GP_19', \
           'G_13', 'G_14', 'G_15', 'G_16', 'G_17', 'G_18', 'G_19', \
           'A_13', 'A_14', 'A_15', 'A_16', 'A_17', 'A_18', 'A_19', \
           'PTS_13', 'PTS_14', 'PTS_15', 'PTS_16', 'PTS_17', 'PTS_18', 'PTS_19', \
           'Sh%_13', 'Sh%_14', 'Sh%_15', 'Sh%_16', 'SH%_17', 'SH%_18', 'SH%_19', \
           'Sh_13', 'Sh_14', 'Sh_15', 'Sh_16', 'iSF_17', 'iSF_18', \
           'TOI/G_13', 'TOI/G_14', 'TOI/G_15', 'TOI/G_16', 'TOI/GP_17', 'TOI/GP_18', 'TOI/GP_19', \
           'TOI/GP_17pp', 'TOI/GP_18pp', 'TOI/GP_19pp', \
           'ZS%_18', 'ZS%_19', \
           'GC_18', 'SCF_17', 'SCF_18', 'GS/G_17', 'GS/G_18', 'GS/G_19', 'OWAR_18', 'OPS_17', 'OPS_18', \
           'PDO_17', 'PDO_18', 'PDO_19', \
           'iCF_17', 'iCF_18', 'iCF_19', \
           'iFF_17', 'iFF_18', \
           'Star_18', \
           'Salary_18', \
           'Cap Cost_13', 'Cap Cost_14', 'Cap Cost_15', 'Cap Hit_16', 'Cap Hit_17', 'Cap Hit_18']]

nhl.rename(columns={'TOI/GP_17': 'TOI/G_17', 'TOI/GP_18': 'TOI/G_18', 'SH%_17': 'Sh%_17', 'SH%_18': 'Sh%_18',
                   'Sh_13': 'iSF_13', 'Sh_14': 'iSF_14', 'Sh_15': 'iSF_15', 'Sh_16': 'iSF_16'}, inplace=True)
nhl.head()

# Rename columns
nhl.rename(columns={'Age_19': 'Age', 'Last Name_19': 'Last_Name', 'First Name_19': 'First_Name', 'Team_19': 'Team',
                    'Position_19': 'Position', 
                    'TOI/GP_17pp': 'PPTOI/GP_17', 'TOI/GP_18pp': 'PPTOI/GP_18', 'TOI/GP_19pp': 'PPTOI/GP_19'}, inplace=True)

nhl.columns
```

### $/PTS

```python
def moneyPerPts(player, pts_year, gp_year, cap_year):
    if player[pts_year] > 0:
        return ((player[gp_year]/82) * player[cap_year]) / player[pts_year]
    else:
        return ((player[gp_year]/82) * player[cap_year])
```

```python
nhl['Cap Cost_13'] = nhl['Cap Cost_13'].replace('[\$,]', '', regex=True).astype(float)
nhl['$/PTS_13'] = nhl.apply(lambda player: moneyPerPts(player, 'PTS_13', 'GP_13', 'Cap Cost_13'), axis=1)

nhl['Cap Cost_14'] = nhl['Cap Cost_14'].replace('[\$,]', '', regex=True).astype(float)
nhl['$/PTS_14'] = nhl.apply(lambda player: moneyPerPts(player, 'PTS_14', 'GP_14', 'Cap Cost_14'), axis=1)

nhl['Cap Cost_15'] = nhl['Cap Cost_15'].replace('[\$,]', '', regex=True).astype(float)
nhl['Cap Cost_15'] = nhl['Cap Cost_15'].apply(lambda player: player*1000000)
nhl['$/PTS_15'] = nhl.apply(lambda player: moneyPerPts(player, 'PTS_15', 'GP_15', 'Cap Cost_15'), axis=1)

nhl['Cap Hit_16'] = nhl['Cap Hit_16'].replace('[\$,]', '', regex=True).astype(float)
nhl['Cap Hit_16'] = nhl['Cap Hit_16'].apply(lambda player: player*1000000)
nhl['$/PTS_16'] = nhl.apply(lambda player: moneyPerPts(player, 'PTS_16', 'GP_16', 'Cap Hit_16'), axis=1)

nhl['Cap Hit_17'] = nhl['Cap Hit_17'].replace('[\$\s,]', '', regex=True).astype(float)
nhl['Cap Hit_17'] = nhl['Cap Hit_17'].apply(lambda player: player/100)
nhl['$/PTS_17'] = nhl.apply(lambda player: moneyPerPts(player, 'PTS_17', 'GP_17', 'Cap Hit_17'), axis=1)

nhl['Cap Hit_18'] = nhl['Cap Hit_18'].replace('[\$\s,]', '', regex=True).astype(float)
nhl['$/PTS_18'] = nhl.apply(lambda player: moneyPerPts(player, 'PTS_18', 'GP_18', 'Cap Hit_18'), axis=1)

nhl['Cap Hit_19'] = nhl['Cap_Hit'].replace('[\$\s,]', '', regex=True).astype(float)
nhl['$/PTS_19'] = nhl.apply(lambda player: moneyPerPts(player, 'PTS_19', 'GP_19', 'Cap Hit_19'), axis=1)
```

### Type Correction

```python
def p2f(x):
    if x == '#DIV/0!':
        return np.nan
    else:
        return float(x.strip('%').replace(',','.'))
```

```python
nhl['ZS%_18'] = nhl[~nhl['ZS%_18'].isna()]['ZS%_18'].apply(lambda feat: p2f(feat)/100)
# nhl['ZS%_19'] = nhl[~nhl['ZS%_19'].isna()]['ZS%_19'].apply(lambda feat: p2f(feat)/100) # Already a float

nhl['GC_18'] = nhl[~nhl['GC_18'].isna()]['GC_18'].apply(lambda feat: p2f(feat))

# nhl['GS/G_19'] = nhl[~nhl['GS/G_19'].isna()]['GS/G_19'].apply(lambda feat: p2f(feat)) # Already a float
nhl['GS/G_18'] = nhl[~nhl['GS/G_18'].isna()]['GS/G_18'].apply(lambda feat: p2f(feat))
nhl['GS/G_17'] = nhl[~nhl['GS/G_17'].isna()]['GS/G_17'].apply(lambda feat: p2f(feat))

nhl['OWAR_18'] = nhl[~nhl['OWAR_18'].isna()]['OWAR_18'].apply(lambda feat: p2f(feat))

nhl['OPS_18'] = nhl[~nhl['OPS_18'].isna()]['OPS_18'].apply(lambda feat: p2f(feat))
nhl['OPS_17'] = nhl[~nhl['OPS_17'].isna()]['OPS_17'].apply(lambda feat: p2f(feat))

nhl['PDO_19'] = nhl[~nhl['PDO_19'].isna()]['PDO_19'].apply(lambda feat: 1-(100-feat))
nhl['PDO_18'] = nhl[~nhl['PDO_18'].isna()]['PDO_18'].apply(lambda feat: 1-p2f(feat))
nhl['PDO_17'] = nhl[~nhl['PDO_17'].isna()]['PDO_17'].apply(lambda feat: 1-p2f(feat))
```

```python
nhl['Sh%_13'] = nhl[~nhl['Sh%_13'].isna()]['Sh%_13'].apply(lambda feat: p2f(feat)/100)
nhl['Sh%_14'] = nhl[~nhl['Sh%_14'].isna()]['Sh%_14'].apply(lambda feat: p2f(feat)/100)
nhl['Sh%_15'] = nhl[~nhl['Sh%_15'].isna()]['Sh%_15'].apply(lambda feat: p2f(feat)/100)
nhl['Sh%_16'] = nhl[~nhl['Sh%_16'].isna()]['Sh%_16'].apply(lambda feat: p2f(feat)/100)

# nhl['Sh%_17'] = nhl['Sh%_17'].replace(',', '.', regex=True)
nhl['Sh%_17'] = nhl[~nhl['Sh%_17'].isna()]['Sh%_17'].apply(lambda feat: p2f(feat)/100)
nhl['Sh%_18'] = nhl['Sh%_18'].replace(',', '.', regex=True)
nhl['Sh%_18'] = nhl[~nhl['Sh%_18'].isna()]['Sh%_18'].apply(lambda feat: p2f(feat)/100)
nhl['SH%_19'] = nhl['SH%_19']/100

nhl['TOI/G_17'] = nhl['TOI/G_17'].replace(',', '.', regex=True).astype(float)
nhl['TOI/G_18'] = nhl['TOI/G_18'].replace(',', '.', regex=True).astype(float)
```

### Variance

```python
def getGPVariance(player):
    GP = []
    # Lock out year
    # if not math.isnan(player.GP_13):
    #   GP.append(player.GP_13)
    if not math.isnan(player.GP_14):
        GP.append(player.GP_14)
    if not math.isnan(player.GP_15):
        GP.append(player.GP_15)
    if not math.isnan(player.GP_16):
        GP.append(player.GP_16)
    if not math.isnan(player.GP_17):
        GP.append(player.GP_17)
    if not math.isnan(player.GP_18):
        GP.append(player.GP_18)
    if not math.isnan(player.GP_19):
        GP.append(player.GP_19)
        
    return np.var(GP)
    
nhl['GP_var'] = nhl.apply(lambda player: getGPVariance(player), axis=1)
```

```python
def getPTSVariance(player):
    PTS = []
    # Lock out year
    # if not math.isnan(player.PTS_13):
    #     PTS.append(player.PTS_13)
    if not math.isnan(player.PTS_14):
        PTS.append(player.PTS_14)
    if not math.isnan(player.PTS_15):
        PTS.append(player.PTS_15)
    if not math.isnan(player.PTS_16):
        PTS.append(player.PTS_16)
    if not math.isnan(player.PTS_17):
        PTS.append(player.PTS_17)
    if not math.isnan(player.PTS_18):
        PTS.append(player.PTS_18)
    if not math.isnan(player.PTS_19):
        PTS.append(player.PTS_19)
        
    return np.var(PTS)
    
nhl['PTS_var'] = nhl.apply(lambda player: getPTSVariance(player), axis=1)
```
